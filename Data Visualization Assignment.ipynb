```python
iris = datasets.load_iris()
```

```python
df=iris.data
```

```python
print(iris.feature_names,iris.target_names)
```

```python
df=pd.DataFrame(data= np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['target'])
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('setosa', fontsize = 15)
ax.set_ylabel('versicolor', fontsize = 15)
ax.set_zlabel('virginica', fontsize = 15)

targets = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']
for target, color,t_name in zip(targets,colors,[0,1,2]):
    indicesToKeep = df[df['target'] == t_name]
    ax.scatter(indicesToKeep['sepal length (cm)'].values, indicesToKeep['sepal width (cm)'].values, 
               indicesToKeep['petal length (cm)'].values,
               c = color, s = 50)
ax.legend(targets)
ax.grid()
plt.show()
```
